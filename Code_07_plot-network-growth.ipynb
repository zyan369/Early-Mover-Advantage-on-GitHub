```
df = pd.read_csv('users.csv', header=None, encoding='utf-8', error_bad_lines=False, warn_bad_lines=False)
df.columns=['user_id', 'login', 'company', 'user_created_at', 'type','fake','deleted', 'long', 'lat', 'country_code','state', 'city', 'location']
df = df.loc[(df['type'] == 'USR') & (df['deleted'] == 0.0) & (df['fake'] == 0)]
```

```
timeline = np.arange(2008, 2018)
count_inset = [np.count_nonzero(pd.to_datetime(df['user_created_at']) < np.datetime64(str(yy+1) + '-01-01')) for yy in timeline ]

fig = plt.figure(figsize =(5.5, 4))
#ax = fig.add_axes([0.1, 0.1, 0.6, 0.75])
#ax.title('Growth of GitHub network size over years')
plt.semilogy(timeline, count_inset, label='Network Size')

#ax.legend(bbox_to_anchor=(1, 1), loc=1, borderaxespad=0.)
#ax.legend('Network Growth', loc=1, borderaxespad=0.)#,bbox_to_anchor=(0.85,0.85))#, prop={'size': 14},fontsize = 'xx-large')
plt.xlabel('Year', fontsize=12)
plt.ylabel('Network Size (number of users)', fontsize=12)
#fig1 = plt.gcf()
#plt.draw()
#ax1 = plt.gca()
#ax1.legend(('y0','y1'))
plt.show()
#ax.xticks(np.arange(2008, 2018, 2))
fig.savefig('networkgrowth.pdf', bbox_inches = 'tight')
```
